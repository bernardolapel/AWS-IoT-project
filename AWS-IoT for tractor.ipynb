```python
from AWSIoTPythonSDK.MQTTLib import AWSIoTMQTTClient
import time
import json
import random

url = 'https://www.amazontrust.com/repository/AmazonRootCA1.pem'

client = AWSIoTMQTTClient("tractor_simulator", useWebsocket=True)
client.configureEndpoint("a38ezfzqzhnmzf-ats.iot.us-east-1.amazonaws.com", 443)
client.configureCredentials("AmazonRootCA1.pem") #, "tractor_simulator.private.key", "tractor_simulator.cert.pem")

client.connect()

 
while True:
    message = {
        "device_id": "tractor_01",
        "gps_lat": round(random.uniform(50.0, 51.0), 6),
        "gps_long": round(random.uniform(-114.0, -113.0), 6),
        "fuel_level": random.randint(50, 100),
        "engine_temp": round(random.uniform(70.0, 95.0), 2),
        "speed_kph": round(random.uniform(5, 25), 2)
    }

    client.publish("farm/tractors/data", json.dumps(message), 1)
    print("Published:", message)
    time.sleep(30)
```
